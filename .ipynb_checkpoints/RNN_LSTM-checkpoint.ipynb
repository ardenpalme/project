## Data Preprocessing and EDA

Check for highly corrlelated features and remove them, you want variables with actual signal.

```python
df1 = df[['Close_NDX', 'Close_^VIX', 'Close_QQQ', 'Close_BTC-USD', 'Close_SOL-USD',
          'Capital Gains','DGS10','GDP','M2V','M2SL','DTWEXBGS','mvrv','nupl']]
df1.corr() # correlation matrix
```

```python
df1 = df1.drop(columns=['Close_NDX'])
```

```python
cols = ['realized_capitalization']
target = ['Close']
skew_cols = ['nupl', 'market_capitalization', 'supply', 'Volume', 'Trade count']

pipe = Pipeline([
    ("preprocessing", ColumnTransformer([
        ("num_scale", StandardScaler(), cols),
        ("tar_scale", StandardScaler(), target),
        ("skew_tfx", PowerTransformer(method='yeo-johnson', standardize=True), skew_cols)
    ],verbose_feature_names_out=False, remainder="passthrough"))
])

display(pipe)
```

```python
data_tfx = pipe.fit_transform(df1)
df2_columns = pipe.get_feature_names_out()
df2 = pd.DataFrame(data_tfx,columns=df2_columns, index=df1.index)
sns.pairplot(df2)
```

```python
X = df2.to_numpy()
y = df2['Close'].to_numpy()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
```

```python
X_train = torch.Tensor(X_train)
y_train = torch.Tensor(y_train)
X_test = torch.Tensor(X_test)
y_test = torwch.Tensor(y_test)
```

## RNN

```python
class RNN(nn.Module):
    def __init__(self, input_size, hidden_size, seq_len):
        super().__init__() # initialize parent class
        self.hidden_size = hidden_size
        self.seq_len = seq_len
        self.rnn_cell = nn.RNNCell(input_size, hidden_size)  
        self.fc = nn.Linear(hidden_size, 1)
        
    def forward(self, x):      
        hx = torch.zeros(self.hidden_size)
        for i in range(self.seq_len):
            hx = self.rnn_cell(x[i], hx)
        return self.fc(hx)  # only last hidden layer to FC layer
```

```python
SEQ_LEN = 40
model = RNN(input_size=X_train.shape[1], hidden_size=16, seq_len=SEQ_LEN)
```

```python
criterion = nn.MSELoss() 
optimizer = optim.Adam(model.parameters(), lr=0.001) 

losses = [] 
for t in range(X_train.shape[0]-SEQ_LEN):
    model.train() 
    output = model(X_train[t:t+SEQ_LEN]) 
    loss = criterion(output.squeeze(), y_train[t+SEQ_LEN])
    
    optimizer.zero_grad() 
    loss.backward()
    optimizer.step() 
    
    losses.append(loss.item())
    if t % 100 == 0:
        print(f"[{t}] loss: {loss.item()}")
```

```python
pred = []

for t in range(X_test.shape[0]-SEQ_LEN):
    model.eval()
    with torch.no_grad():
        pred.append(model(X_test[t:t+SEQ_LEN]).squeeze().numpy())
pred = np.array(pred)

tar_scale = pipe.named_steps["preprocessing"].named_transformers_["tar_scale"]
y_test_inv = tar_scale.inverse_transform(y_test.reshape(-1,1)).T[0]
pred_inv = tar_scale.inverse_transform(pred.reshape(-1,1)).T[0]
```

```python
ser1 = pd.Series(pred_inv, index=df2.index[-(y_test.shape[0]-SEQ_LEN):])
ser2 = pd.Series(df1['Close'].values[:-(y_test.shape[0]-SEQ_LEN)], index=df1.index[:-(y_test.shape[0]-SEQ_LEN)])
ser3 = pd.Series(y_test_inv[-(y_test.shape[0]-SEQ_LEN):], index=df2.index[-(y_test.shape[0]-SEQ_LEN):])
ser1.index = pd.to_datetime(ser1.index)
ser2.index = pd.to_datetime(ser2.index)
ser3.index = pd.to_datetime(ser3.index)

losses_ser = pd.Series(losses, index=df1.index[SEQ_LEN:X_train.shape[0]])
losses_ser.index = pd.to_datetime(losses_ser.index)

fig, axs = plt.subplots(1,2, figsize=(12, 6))
axs[0].plot(losses_ser)

axs[1].plot(ser2, color = "orange", label="close")
axs[1].plot(ser1, color="red", linestyle="dashed", label="predicted")
axs[1].plot(ser3, color="blue", label="actual")
axs[1].axvline(x=ser3.index[0], color='red', linestyle=':', alpha=0.7, label='Prediction Start')
axs[1].legend()

for ax in axs:
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
    ax.xaxis.set_major_locator(mdates.AutoDateLocator(maxticks=12))
    ax.tick_params(axis='x', rotation=45)

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## LSTM

```python
class LSTM(nn.Module):
    def __init__(self, input_size, hidden_size, seq_len):
        super().__init__()
        self.hidden_size = hidden_size
        self.seq_len = seq_len
        self.lstm_cell = nn.LSTMCell(input_size, hidden_size)
        self.fc = nn.Linear(hidden_size, 1)

    def forward(self, x):
        hx = torch.zeros(self.hidden_size)
        cx = torch.zeros(self.hidden_size)
        for i in range(self.seq_len):
            hx, cx = self.lstm_cell(x[i], (hx,cx))
        return self.fc(hx)
```

```python
SEQ_LEN = 10
model = LSTM(input_size=X_train.shape[1], hidden_size=12, seq_len=SEQ_LEN)
```

```python
criterion = nn.MSELoss() 
optimizer = optim.SGD(model.parameters(), lr=0.01) 

losses = [] 
for t in range(X_train.shape[0]-SEQ_LEN):
    model.train() 
    output = model(X_train[t:t+SEQ_LEN]) 
    loss = criterion(output.squeeze(), y_train[t+SEQ_LEN])
    
    optimizer.zero_grad() 
    loss.backward()
    optimizer.step() 
    
    losses.append(loss.item())
    if t % 100 == 0:
        print(f"[{t}] loss: {loss.item()}")
```

```python
pred = []

for t in range(X_test.shape[0]-SEQ_LEN):
    model.eval()
    with torch.no_grad():
        pred.append(model(X_test[t:t+SEQ_LEN]).squeeze().numpy())
pred = np.array(pred)

tar_scale = pipe.named_steps["preprocessing"].named_transformers_["tar_scale"]
y_test_inv = tar_scale.inverse_transform(y_test.reshape(-1,1)).T[0]
pred_inv = tar_scale.inverse_transform(pred.reshape(-1,1)).T[0]
```

```python
ser1 = pd.Series(pred_inv, index=df2.index[-(y_test.shape[0]-SEQ_LEN):])
ser2 = pd.Series(df1['Close'].values[:-(y_test.shape[0]-SEQ_LEN)], index=df1.index[:-(y_test.shape[0]-SEQ_LEN)])
ser3 = pd.Series(y_test_inv[-(y_test.shape[0]-SEQ_LEN):], index=df2.index[-(y_test.shape[0]-SEQ_LEN):])
ser1.index = pd.to_datetime(ser1.index)
ser2.index = pd.to_datetime(ser2.index)
ser3.index = pd.to_datetime(ser3.index)

losses_ser = pd.Series(losses, index=df1.index[SEQ_LEN:X_train.shape[0]])
losses_ser.index = pd.to_datetime(losses_ser.index)

fig, axs = plt.subplots(1,2, figsize=(12, 6))
axs[0].plot(losses_ser)

axs[1].plot(ser2, color = "orange", label="close")
axs[1].plot(ser1, color="red", linestyle="dashed", label="predicted")
axs[1].plot(ser3, color="blue", label="actual")
axs[1].axvline(x=ser3.index[0], color='red', alpha=0.7, label='Prediction Start')
axs[1].legend()

for ax in axs:
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
    ax.xaxis.set_major_locator(mdates.AutoDateLocator(maxticks=12))
    ax.tick_params(axis='x', rotation=45)

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
